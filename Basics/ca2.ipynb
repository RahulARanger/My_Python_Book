> ### Question: 1

<img src='https://user-images.githubusercontent.com/57987728/91119319-82202a00-e6b0-11ea-9745-126a766f294f.PNG' >

```python
import math
def factorial(n):
    if n==0 or n==1:return 1
    else:return n*factorial(n-1)
def calc(x,n):
    check=0
    for i in range(n):
        odd=2*i+1
        new=x**(odd)
        new/=factorial(odd)
        new*=(-1)**i
        check+=new
    return float(check)
while True:
    a=input('1 for input through degrees and 2 for radians: ')
    if a=='1':
        x=float(input('x (in degrees): '))
        x_=math.radians(x)
        break
    elif a=='2':
        x_=float(input('x (in radians): '))
        x=math.degrees(x_)
        break
    else:print('Please Try again')
# recommended n=20
n=int(input('n: '))
print('sin({} deg): {:.3f}'.format(x,calc(x_,n)))
```
