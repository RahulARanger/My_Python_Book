# MIME Types

Jupyter Notebook can also display many other **MIME**_(MultiPurpose Internet Mail Extensions)_ types not just text.

for that we need to import some IPython packages, **Note:** This modules won't work inside a pure python scripts

## Displaying Images

To display an image in IPython, we need to import `SVG`, `Image`, `display` from `IPython.display`.

`Image` object can deal with most of the image formats.
`display` function is used to display `Image` object

`SVG` object is used for the _SVG_ files, `display` function also display `SVG` objects

```python
from IPython.display import Image, display

Image("Rem.jpg")  # display JPG or PNG images
```

```python
display(Image("rem.jpg"))   # displays images 
```

```python
# can also request for the images from url
Image(url="https://c.tenor.com/1-i2wuiJspUAAAAM/rem.gif")  # same as Image("https://c.tenor.com/1-i2wuiJspUAAAAM/rem.gif")
```

```python
from IPython.display import SVG

orange = SVG("orange.svg")
display(orange)
```

```python
from IPython.display import HTML

# we also have magic method (cell magic) which can do this
sample = """
<table>

<thread>
    
    <tr>
    <th> Name </th>
    <th> version </th>
    </tr>
    
</thead>


<tbody>
    <tr>
    
    <td> Rash </td>
    <td> 0.0.5 </td>
    
    </tr>
    
    <tr>
    
    <td> RashLogger </td>
    <td> 0.0.1 </td>
    
    </tr>
    
    <tr>
    
    <td> RashNotify </td>
    <td> 0.0.0 </td>
    
    </tr>

</tbody>

</table>

"""

display(HTML(sample))
```

We can also display Javascript, inside Jupyter Notebook, but notebook becomes **Not Trusted**, since it can execute any Javascript, it might contain harmful code. 
at the top right corner, we can whether a notebook is **Trusted** or not.
It also has it's magic method `%%javascript`

```python
from IPython.display import Javascript

sample = """
alert("Hello")
"""
display(Javascript(sample))
```
